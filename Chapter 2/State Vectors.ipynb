```python
import qiskit
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from math import sqrt, pi

#linear combinations describe 'superposition' 
#To find the probability of measuring a state  |ψ⟩  in the state  |x⟩  we do: p(|x⟩)=|⟨x|ψ⟩|2
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
circuit = qiskit.QuantumCircuit(1)
inital = [1/sqrt(3), sqrt(2/3)]
#inital = [1/sqrt(2), 1j/sqrt(2)]

circuit.initialize(inital,0)

circuit.draw(output='mpl')
```

```python
counts = execute(circuit,backend).result().get_counts()
plot_histogram(counts)
```

```python
for i in range(10):
    result = execute(circuit,backend).result()
    stateVector = result.get_statevector()
    if (stateVector == [0.+1.j, 0.+0.j]).all():
        print("|0>")
    elif (stateVector == [0.+0.j, 1.+0.j]).all():
        print("|1>")
```
