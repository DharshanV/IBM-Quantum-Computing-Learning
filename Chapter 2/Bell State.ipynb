```python
import qiskit
from qiskit import QuantumCircuit, Aer, execute, ClassicalRegister
from qiskit.visualization import plot_bloch_multivector,plot_histogram
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
circuit = QuantumCircuit(2)

initalQbit0 = [1,0]
initalQbit1 = [1,0]

circuit.initialize(initalQbit0,0)
circuit.initialize(initalQbit1,1)

circuit.barrier()
circuit.h(0)
circuit.cx(0,1)
circuit.barrier()

circuit.draw('mpl')
```

```python
result = execute(circuit,backend).result()
counts = result.get_counts()
plot_histogram(counts)                      #Since the two bits entangled. The measument of one tells the value of the other, since its 50/50.
```
